# Model Selection

```python
#1->Split the data
#the training set used for training model and test set used for assess the performance of the model

#settings a random seed so our model gives fixed results
np.random.seed(42)
```

```python
#feature variable without Revenue column because target variable is Revenue
x = data.drop('Revenue', axis=1)
x.head(2)
```

```python
x.shape
```

```python
y = data['Revenue']
y.head()
```

## Split Data

```python
#Split the data to train and test test

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4)  #test_size mean split to dataset 60%-training set and 40%-testing set

#checking the dimentions of the train and test dataset
print(f"x_train:{x_train.shape} , x_test:{x_test.shape} y_train:{y_train.shape} , y_test:{y_test.shape}")
```

```python
#Train the model using Linear Regression

from sklearn.linear_model import LinearRegression

model = LinearRegression()

#fit the model
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

# EVALUATING THE MODEL

## Evaluating Linear Regression model

```python
"""
Evaluating methods-->(Evaluation Metrix)
    1-MSE
    2-R2 Score
    3-MAE
"""
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

mse = mean_squared_error(y_test, y_pred)
r2Score = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print("Mean Squared Error: ", mse)
print("R Squared Error: ", r2Score)
print("Mean Absolute Error: ", mae)

print("Train Accuracy: ", model.score(x_train,y_train))
print("Test Accuracy",model.score(x_test, y_test))
```

```python
#visualizing the actual vs predicted values
plt.scatter(y_pred, y_test, color='blue')
plt.xlabel("Predicted values")
plt.ylabel("Actual values")
plt.title("Actual Vs Predicted Values")
plt.xticks(rotation=30) #rotate x variables
plt.show()
```

```python
#ploting 3D visualization

from sklearn.decomposition import PCA
#3D visualization (using PCA)
pca = PCA(n_components= 3)
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.fit_transform(x_test)

#Plotting actual vs predicted values in 3D space
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

#scatter plot of actual values
ax.scatter(x_test_pca[:, 0], x_test_pca[:, 1], y_test, color='blue', label='Actual')

#scatter plot of predicted values
ax.scatter(x_test_pca[:, 0], x_test_pca[:, 1], y_pred, color='red', label='Predicted') #we use the scatter plot for visualizing 3D plot using PCA

#set labels and tittle
ax.set_xlabel('Component 1')
ax.set_ylabel('Component 2')
ax.set_zlabel('Y')
ax.set_title('3D plot to Linear Regression Actual Vs Predicted Values')

plt.show()
```

```python
#============================================#
# Create model Using Random Forest Regressor
#============================================#
```

```python
from sklearn.ensemble import RandomForestRegressor #applying the random forest regressor
clf = RandomForestRegressor()
#fit the regressor model
clf.fit(x_train, y_train)
```

```python
#prediction using predicted data set
y2_pred = clf.predict(x_test)
```

# Evaluating RandomForestRegressor

```python
"""
Evaluating methods-->
    1-MSE
    2-R2 Score
    3-MAE

    clf mean equal to random forest regressor
"""
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

mse = mean_squared_error(y_test, y2_pred)
r2Score = r2_score(y_test, y2_pred)
mae = mean_absolute_error(y_test, y2_pred)

print("Mean Squared Error: ", mse)
print("R Squared Error: ", r2Score)
print("Mean Absolute Error: ", mae)
print("Training Accuracy: ", clf.score(x_train,y_train))
print("Testing Acuracy: ", clf.score(x_test,y_test))
```

```python
#Visualize the outcoe
plt.scatter(y_test, y2_pred, color = 'gray')
plt.xlabel("Predicted values")
plt.ylabel("Actual values")
plt.title("Actual Vs Predicted Values")
plt.show()
```

```python
#ploting 3D visualization

from sklearn.decomposition import PCA
#3Dr visualization (using PCA)
pca = PCA(n_components= 3)
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.fit_transform(x_test)

#Plotting actual vs predicted values in 3D space
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

#scatter plot of actual values
ax.scatter(x_test_pca[:, 0], x_test_pca[:, 1], y_test, color='red', label='Actual')

#scatter plot of predicted values
ax.scatter(x_test_pca[:, 0], x_test_pca[:, 1], y2_pred, color='blue', label='Predicted') #we use the scatter plot for visualizing 3D plot using PCA

#set labels and tittle
ax.set_xlabel('Component 1')
ax.set_ylabel('Component 2')
ax.set_zlabel('Y')
ax.set_title('3D plot to Linear Regression Actual Vs Predicted Values')

plt.show()
```
